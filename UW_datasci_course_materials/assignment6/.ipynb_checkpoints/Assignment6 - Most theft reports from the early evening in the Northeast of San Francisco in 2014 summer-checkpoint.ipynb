```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
pd.options.mode.chained_assignment = None 
import numpy as np
import seaborn as sns; sns.set()
from operator import add
```

```python
#########################
## Import the datafile
#########################

sf_crime = pd.read_csv('sanfrancisco_incidents_summer_2014.csv', infer_datetime_format=True)

# Show some data
print sf_crime.shape # About 29k records in the data
print sf_crime.head(1)
```

```python
# Confirm the data that is from summer 2014
date=pd.to_datetime(sf_crime['Date'])
print date.min()
print date.max()
```

```python
###################################################
## Theft including the Vehicle Theft are the most 
## crimes in the San Francisco in summer 2014
###################################################

# Count the percent of each category 
Category = pd.DataFrame()
Category['Category'] = sorted(sf_crime['Category'].unique())
Category['Count_pct'] = list(sf_crime['Category'].groupby(sf_crime['Category']).agg('count') / float(len(sf_crime)))
Category = Category.sort_values('Count_pct', ascending=False)

#######################################
## Keep top five categories and combine 
## the rest categories to 'OTHERS'
#######################################

# Category labels
Category_labels = list(Category['Category'].head(5))
Category_labels.insert(1, 'OTHERS')
# Category sizes
Category_sizes = list(Category['Count_pct'].head(5))
Category_sizes.insert(1, 1-sum(Category['Count_pct'].head(5)))

# Plot Category pct pie chart
fig = plt.figure()
fig.set_size_inches(6, 6)

pie = plt.pie(Category_sizes, explode=(0.1, 0, 0, 0, 0, 0.1), 
              labels=Category_labels, autopct='%1.1f%%',
        shadow=False, startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Figure 1 - Crime Percent by Category', y=1.1, fontsize=15)
plt.tight_layout()
```

### Figure 1 shows that theft including vehicle theft is the top crime (39.4% total) in SF 2014 summer.

```python
#############################################
## Extract 'THEFT' report from big dataset
#############################################

Theft = sf_crime[(sf_crime['Category'].isin(['LARCENY/THEFT','VEHICLE THEFT']))]

# Convert 'Time' to datetime hourly
Theft.loc[:,'Time2'] = pd.to_datetime(Theft.Time).dt.hour
```

```python
#############################################
## Draw theft reports as dot on SF map
#############################################
# The codes from this cell is adapted from 
# http://nbviewer.jupyter.org/github/lmart999/GIS/blob/master/SF_GIS_Crime.ipynb

from mpl_toolkits.basemap import Basemap
import fiona
from shapely.geometry import shape, mapping
from pyproj import Proj, transform
from fiona.crs import from_epsg
from itertools import chain

shapefile="/Users/Linghuan/Documents/Coursera/datasci_course_materials/assignment6/sfcoord2nbh-master/sffind_neighborhoods/SFFind_neighborhoods.shp"
shp = fiona.open(shapefile)
bds = shp.bounds

shp.close()
extra = 0.01
ll = (bds[0], bds[1])
ur = (bds[2], bds[3])
coords = list(chain(ll, ur))
w, h = coords[2] - coords[0], coords[3] - coords[1]

m = Basemap(
    projection='tmerc',
    lon_0=-122.,
    lat_0=37.7,
    ellps = 'WGS84',
    llcrnrlon=coords[0] - extra * w,
    llcrnrlat=coords[1] - extra + 0.01 * h,
    urcrnrlon=coords[2] + extra * w,
    urcrnrlat=coords[3] + extra + 0.01 * h,
    lat_ts=0,
    resolution='i',
    suppress_ticks=True)

m.readshapefile(
    '/Users/Linghuan/Documents/Coursera/datasci_course_materials/assignment6/sfcoord2nbh-master/sffind_neighborhoods/SFFind_neighborhoods',
    'SF',
    color='none',
    zorder=2)

from descartes import PolygonPatch
from shapely.geometry import Point, Polygon, MultiPoint, MultiPolygon
from shapely.prepared import prep
from matplotlib.collections import PatchCollection

# Set up a map dataframe
df_map = pd.DataFrame({
    'poly': [Polygon(xy) for xy in m.SF],
    'ward_name': [ward['name'] for ward in m.SF_info]})
df_map['area_m'] = df_map['poly'].map(lambda x: x.area)
df_map['area_km'] = df_map['area_m'] / 100000

def makePoints(dat):
    # Create Point objects in map coordinates from dataframe lon and lat values
    map_points = pd.Series([Point(m(mapped_x,mapped_y)) for mapped_x, mapped_y in zip(dat['X'],dat['Y'])]) # X: longitude; Y: latitude 
    plt_points = MultiPoint(list(map_points.values))
    hoods_polygon = prep(MultiPolygon(list(df_map['poly'].values)))
    pts = filter(hoods_polygon.contains,plt_points)
    return pts

sf_Theft_points_todraw=makePoints(Theft)


# Draw neighborhoods withpolygons
df_map['patches'] = df_map['poly'].map(lambda x: PolygonPatch(
    x,
    fc='#000000',
    ec='red', 
    lw=.5, alpha=0.2,
    zorder=4))

plt.clf()
fig = plt.figure()
fig.set_size_inches(8,12)
ax = fig.add_subplot(111, axisbg='w', frame_on=False)

# Now, we can overlay our points
dev = m.scatter(
    [geom.x for geom in sf_Theft_points_todraw],
    [geom.y for geom in sf_Theft_points_todraw],
    10, marker='o', lw=.25,
    facecolor='red', edgecolor='cyan',
    alpha=0.75, antialiased=True,
    label='Theft', zorder=3)

ax.add_collection(PatchCollection(df_map['patches'].values, match_original=True))
plt.title('Figure 2 - San Francisco Theft Reports in Summer 2014', fontsize=15, y=0.95)
plt.tight_layout()
```

### Figure 2 shows that as we expected, theft is more common in the residential areas, and is uncommon in recreation park areas, such as Presidio park and Golden State Park.  Most thefts occurred in the northeast of San Francisco, including Northern, Southern, Central and Tenderloin districts. The other districts have much less theft reports. 

```python
######################################################################
## Theft Clustermap of PdDistrict and theft time (only kept in hour)
######################################################################

# Adapt from http://nbviewer.jupyter.org/github/lmart999/GIS/blob/master/SF_GIS_Crime.ipynb
def types(d_crime,per,groupby_cate):
    # Group by crime type and district 
    hoods_per_type=d_crime.groupby(groupby_cate).PdDistrict.value_counts(sort=True)
    t=hoods_per_type.unstack().fillna(0)
    
    # Sort by hood sum
    hood_sum=t.sum(axis=0)
    hood_sum.sort_values(ascending=False)
    t=t[hood_sum.index]
    
    # Filter by crime per district
    crime_sum=t.sum(axis=1)
    crime_sum.sort_values()
    
    # Large number, so let's slice the data.
    p=np.percentile(crime_sum,per)
    ix=crime_sum[crime_sum>p]
    t=t.loc[ix.index]
    return t

# Count the theft based on the time - hourly
Theft_time = types(Theft, 0, 'Time')
```

```python
# Clustermap of Theft
theft_clustermap = sns.clustermap(Theft_time, standard_scale=1, linewidths=.5) 
plt.setp(theft_clustermap.ax_heatmap.yaxis.get_majorticklabels(), rotation=0)
plt.title('Figure 3 - Clustermap of Police Deparment District and Theft Time (Hour)', fontsize=15, x=13, y=1.3)
plt.text(14.2, -6.5, "", rotation='horizontal')
```

### Figure 3 shows that most thefts occurred in the early evening (17:00 - 20:00) in all the police deparment districts. 18:00 is the peak hour of the theft reports across most districts.

```python
##############################
## Theft Report Histogram
##############################

# Calculate the theft report percent on each PdDistrict
Theft_group = pd.DataFrame({'Percent': Theft.groupby(['PdDistrict','Time']).size()/float(len(Theft))}).reset_index()

# Bar hours - 24 hours reference
bar_hr = xrange(24)

# X axis label - order from 5:00 - 4:00
xlabels = ["%i:00" % i for i in range(5, 24)] + ["%i:00" % i for i in range(5)]

# Bar lists that contains all the bars statistics
bar_lists = list()

# Contain the bar information
bars = range(len(bar_hr))

# Specify the color of bars
color = ['white', 'blue', 'brown', 'purple', 'pink', 'grey', 'green', 'yellow', 'navy', 'red']

# Calculate the bar percent based on PdDistrict    
def bar_pct(district):    
    
    # Get the specified PdDistrict theft data
    Theft_bar = Theft_group[Theft_group['PdDistrict'] == district][['Time','Percent']]
    
    # Check whether has 24 hours data
    for i in xrange(len(bar_hr)):
        if Theft_bar.iloc[i]['Time'] != i:
            Theft_bar.loc[i] = [i, 0] # insert the missing time (hour)
            Theft_bar = Theft_bar.sort_values('Time') # Sort by 'Time'
            Theft_bar = Theft_bar.set_index(Theft_bar['Time']) # reset the DataFrame index
            
    # Reorder the bar pct data from 5 to 4
    Theft_bar_pct = list(Theft_bar['Percent'])[5:]+list(Theft_bar['Percent'][:5])
    
    # Accumulate the bar statistics for current bar
    if len(bar_lists) != 0:
        Theft_bar_pct = map(add, Theft_bar_pct, bar_lists[len(bar_lists)-1])
    
    # Append to bar_lists
    bar_lists.append(Theft_bar_pct)
    
    return bar_lists

# Dar_bar function
def draw_bar(bar_lists, bars, color): 
    
    # Drawing a stacked bar must start from the farthest bar to the nearest bar. The largest percent bar come first
    for i in xrange(len(bar_lists)-1, -1, -1):
        # Create the bars
        bars[i] = plt.bar(bar_hr, bar_lists[i], align='center')

        # Change the bar color
        for j in xrange(len(bar_hr)):    
            bars[i][j].set_color(color[i])
            # the shade in the color shows the amount of the reports
            bars[i][j].set_alpha(bar_lists[i][j]/max(bar_lists[i])) 
    return
            
# Prepare bar statistics 
# The bar order is determined by the percent. The smallest comes first.
bar_lists = bar_pct(district ='TENDERLOIN') # TENDERLOIN District Police Department
bar_lists = bar_pct(district='TARAVAL') # TARAVAL District Police Department
bar_lists = bar_pct(district ='PARK') # PARK District Police Department
bar_lists = bar_pct(district='RICHMOND') # RICHMOND District Police Department
bar_lists = bar_pct(district ='BAYVIEW') # BAYVIEW District Police Department
bar_lists = bar_pct(district='MISSION') # MISSION District Police Department
bar_lists = bar_pct(district ='INGLESIDE') # INGLESIDE District Police Department
bar_lists = bar_pct(district='NORTHERN') # NORTHERN District Police Department
bar_lists = bar_pct(district ='CENTRAL') # CENTRAL District Police Department
bar_lists = bar_pct(district='SOUTHERN') # SOUTHERN District Police Department

# Set the figure size
plt.figure(figsize=(16, 8))

# Draw the bars
draw_bar(bar_lists, bars, color)

# Find all the peak points on the highest bar
peak_pts = list()
for i in (range(1,len(bar_lists[len(bar_lists)-1])-1)): 
    if bar_lists[len(bar_lists)-1][i] > bar_lists[len(bar_lists)-1][i-1] \
    and bar_lists[len(bar_lists)-1][i] > bar_lists[len(bar_lists)-1][i+1]:      
        peak_pts.append(i)

# Find the max peak pt on the highest bar   
peak_pt_max = list()
for pt in peak_pts:
    if bar_lists[len(bar_lists)-1][pt] == max([bar_lists[len(bar_lists)-1][pt] for pt in peak_pts]):
        peak_pt_max.append(pt)       

# Plot the max point
plt.scatter(peak_pt_max[0], bar_lists[len(bar_lists)-1][peak_pt_max[0]]+0.0025, c='magenta', s=800, marker="*")

# Plot the trend arrow lines
plt.arrow(0,0.015,13,0.085, length_includes_head=True, head_width=0.006, head_length=0.6, color='red')
plt.arrow(13.5,0.1,10,-0.085, length_includes_head=True, head_width=0.006, head_length=0.6, color='orange')

# Customize the x ticks values and label
plt.xticks(bar_hr, xlabels, fontsize=14)
plt.xlabel('Time', fontsize=16)


## Create the % label for y axis
yticks = [i*.01 for i in range(0, 12, 2)]
ylabels = ["%i%%" % i for i in range(0, 12, 2)]

# Customize the y ticks values and label
plt.ylabel('Percent', fontsize=16)
plt.yticks(yticks, ylabels, fontsize=14)

# Create title
plt.title('Figure 4 - Time (Hour) Percent of Theft Reports', fontsize=15, y=1.05)

# Grow the y axis 
plt.ylim(0,0.105)
# Expand the x axis 
plt.xlim(-1, 24) 

# Plot the legend
plt.legend( #Specify the bars for the legend colors 
           (bars[9][13], bars[8][13], bars[7][13],
            bars[6][13], bars[5][13], bars[4][13],
            bars[3][13], bars[2][13], bars[1][13],
            bars[0][13] ),
            # Specify the legend values
           ('SOUTHERN', 'CENTRAL', 'NORTHERN',
            'INGLESIDE', 'MISSION', 'BAYVIEW', 
            'RICHMOND', 'PARK', 'TARAVAL',
            'TERDERLOIN'), loc='best')        
        
```

### Figure 4 confirms that in 2014 summer San Francisco Northeast area (including Sourthern, Central and Northern districts) has the most theft reports, especially from 17:00 to 20:00 daily with peak at 18:00. Tenderloin district, also in the Northeast of San Francisco, has relatively low percentage compared to other three Northeast districts. Probably that is due to the small area of Tenderloin, which is surrounded by the other districts.  In addition, the theft report has an upward trend from 5:00 to 18:00, and a downward trend from 18:00 to 4:00.
