```python
def types_districts(d_crime,per,groupby_cate):
    # Group by crime type and district 
    hoods_per_type=d_crime.groupby(groupby_cate).PdDistrict.value_counts(sort=True)
    t=hoods_per_type.unstack().fillna(0)
    
    # Sort by hood sum
    hood_sum=t.sum(axis=0)
    hood_sum.sort(ascending=False)
    t=t[hood_sum.index]
    
    # Filter by crime per district
    crime_sum=t.sum(axis=1)
    crime_sum.sort()
    
    # Large number, so let's slice the data.
    p=np.percentile(crime_sum,per)
    ix=crime_sum[crime_sum>p]
    t=t.loc[ix.index]
    return t
    
sf_crime_districts = types_districts(sf_crime,0, 'Category')
```

```python
# Hourly bar Pct
hour_bar_pct = Theft_group_hr['Hour'].agg('count') / float(len(Theft))       

# Hourly bar hour
hour_bar_hr = range(24)    
   
# Create the bars
barlist = plt.bar(hour_bar_hr, hour_bar_pct, align='center', alpha=0.5)
# Change the bar color
for i in range(len(hour_bar_pct)):    
    barlist[i].set_color((1, 1-hour_bar_pct[i]/max(hour_bar_pct), 0))    

# Find all the peak pts    
peak_pts = []
for i in (range(1,len(hour_bar_pct)-1)):
    if hour_bar_pct[i] > hour_bar_pct[i-1] \
    and hour_bar_pct[i] > hour_bar_pct[i+1]:
        peak_pts.append(i)

# Plot the peak pts        
for i in range(len(peak_pts)):
    plt.scatter(hour_bar_hr[peak_pts[i]], hour_bar_pct[peak_pts[i]], c='magenta', s =500, marker = "*")
        
# Plot the trendline
plt.plot(hour_bar_pct, c='sandybrown', linewidth=1)

# Plot the arrows
plt.arrow(0,0.045,4.5,-0.035, length_includes_head=True,head_width=0.006, head_length=0.6, color = 'orange')
plt.arrow(5,0.015,7.5,0.05, length_includes_head=True,head_width=0.006, head_length=0.6, color = 'red')
plt.arrow(13,0.05,5.5,0.05, length_includes_head=True,head_width=0.006, head_length=0.6, color = 'red')
plt.arrow(19,0.085,4.5,-0.025, length_includes_head=True,head_width=0.006, head_length=0.6, color = 'orange')

# Customize the x ticks
plt.xticks(hour_bar_hr, hour_bar_hr, fontsize=13)

# Customize x label
plt.xlabel('Time', fontsize=15)

# Customize y label
plt.ylabel('Percent', fontsize=15)

# Create the title
plt.title('Hourly Percent of Theft Reports', fontsize=15)

# Grow the y axis
plt.ylim(0,0.105)
# Expand the x axis
plt.xlim(-1, 24)    
```

```python
from mpl_toolkits.basemap import Basemap
import fiona
from shapely.geometry import shape, mapping
from pyproj import Proj, transform
from fiona.crs import from_epsg
from itertools import chain

shapefile="/Users/Linghuan/Documents/Coursera/datasci_course_materials/assignment6/sfcoord2nbh-master/sffind_neighborhoods/SFFind_neighborhoods.shp"
shp = fiona.open(shapefile)
bds = shp.bounds
print (bds)

shp.close()
extra = 0.01
ll = (bds[0], bds[1])
ur = (bds[2], bds[3])
coords = list(chain(ll, ur))
w, h = coords[2] - coords[0], coords[3] - coords[1]
```

```python
m = Basemap(
    projection='tmerc',
    lon_0=-122.,
    lat_0=37.7,
    ellps = 'WGS84',
    llcrnrlon=coords[0] - extra * w,
    llcrnrlat=coords[1] - extra + 0.01 * h,
    urcrnrlon=coords[2] + extra * w,
    urcrnrlat=coords[3] + extra + 0.01 * h,
    lat_ts=0,
    resolution='i',
    suppress_ticks=True)

m.readshapefile(
    '/Users/Linghuan/Documents/Coursera/datasci_course_materials/assignment6/sfcoord2nbh-master/sffind_neighborhoods/SFFind_neighborhoods',
    'SF',
    color='none',
    zorder=2)
```

```python
from descartes import PolygonPatch
from shapely.geometry import Point, Polygon, MultiPoint, MultiPolygon
from shapely.prepared import prep
from matplotlib.collections import PatchCollection

# Set up a map dataframe
df_map = pd.DataFrame({
    'poly': [Polygon(xy) for xy in m.SF],
    'ward_name': [ward['name'] for ward in m.SF_info]})
df_map['area_m'] = df_map['poly'].map(lambda x: x.area)
df_map['area_km'] = df_map['area_m'] / 100000

def makePoints(dat):
    # Create Point objects in map coordinates from dataframe lon and lat values
    map_points = pd.Series([Point(m(mapped_x,mapped_y)) for mapped_x, mapped_y in zip(dat['X'],dat['Y'])]) # X: longitude; Y: latitude 
    plt_points = MultiPoint(list(map_points.values))
    hoods_polygon = prep(MultiPolygon(list(df_map['poly'].values)))
    pts = filter(hoods_polygon.contains,plt_points)
    return pts

sf_Theft_Solved_points_todraw=makePoints(Theft[Theft['Resolution2'] == 'Solved'])
sf_Theft_Unsolved_points_todraw=makePoints(Theft[Theft['Resolution2'] == 'Unsolved'])
```

```python
# Draw neighborhoods withpolygons
df_map['patches'] = df_map['poly'].map(lambda x: PolygonPatch(
    x,
    fc='#000000',
    ec='blue', 
    lw=.5, alpha=0.2,
    zorder=4))

plt.clf()
fig = plt.figure()
ax = fig.add_subplot(111, axisbg='w', frame_on=False)

# Now, we can overlay our points
dev = m.scatter(
    [geom.x for geom in sf_Theft_Unsolved_points_todraw],
    [geom.y for geom in sf_Theft_Unsolved_points_todraw],
    10, marker='o', lw=.25,
    facecolor='red', edgecolor='cyan',
    alpha=0.75, antialiased=True,
    label='Theft', zorder=3)


dev = m.scatter(
    [geom.x for geom in sf_Theft_Solved_points_todraw],
    [geom.y for geom in sf_Theft_Solved_points_todraw],
    10, marker='o', lw=.25,
    facecolor='green', edgecolor='cyan',
    alpha=0.75, antialiased=True,
    label='Theft', zorder=3)



ax.add_collection(PatchCollection(df_map['patches'].values, match_original=True))
plt.tight_layout()
fig.set_size_inches(15,15)
```
