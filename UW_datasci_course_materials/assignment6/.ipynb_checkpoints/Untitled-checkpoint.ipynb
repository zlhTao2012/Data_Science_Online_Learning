```python
def plotdat(data,cat):
    l=data.groupby(cat).size()
    l.sort()
    fig=plt.figure(figsize=(10,5))
    plt.yticks(fontsize=8)
    l.plot(kind='bar',fontsize=12,color='k')
    plt.xlabel('')
    plt.ylabel('Number of Reports',fontsize=10)

plotdat(sf_crime,'Category')
```

```python
def types_districts(d_crime,per,groupby_cate):
    # Group by crime type and district 
    hoods_per_type=d_crime.groupby(groupby_cate).PdDistrict.value_counts(sort=True)
    t=hoods_per_type.unstack().fillna(0)
    
    # Sort by hood sum
    hood_sum=t.sum(axis=0)
    hood_sum.sort(ascending=False)
    t=t[hood_sum.index]
    
    # Filter by crime per district
    crime_sum=t.sum(axis=1)
    crime_sum.sort()
    
    # Large number, so let's slice the data.
    p=np.percentile(crime_sum,per)
    ix=crime_sum[crime_sum>p]
    t=t.loc[ix.index]
    return t
    
t=types_districts(sf_crime,0, 'Category')
```

```python
# Clustermap
cg = sns.clustermap(t, linewidths=.5) 
# Rotate clustermap labels
plt.setp(cg.ax_heatmap.yaxis.get_majorticklabels(), rotation=0)
```

```python
# Extract 'THEFT' report from big dataset
Theft = sf_crime[(sf_crime['Category'].isin(['LARCENY/THEFT','VEHICLE THEFT']))]
Theft.loc[:,'Cate_new'] = 'THEFT'
```

```python
t=types_districts(Theft,0, 'Descript')
# Clustermap of Theft
cg_theft = sns.clustermap(t, linewidths=.5) 
plt.setp(cg_theft.ax_heatmap.yaxis.get_majorticklabels(), rotation=0)
```

```python
# Create 'Resolution2' label by Assigning 'None' to 'NONE' or 'UNFOUNDED' on 'Resolution' label, 'Solved' to others
Theft.loc[:,'Resolution2'] = ['Unsolved' if event in ['NONE', 'UNFOUNDED'] else 'Solved' for event in Theft.loc[:,'Resolution']]

# Based on judgement call, assign 'Short' or 'Long' to newly created 'Report' label
# 'Short' means the time between theft occurred and theft report is short. 
Theft.loc[:,'Report'] = ['Short' if event in [        

'ATTEMPTED SHOPLIFTING',
'ATTEMPTED THEFT FROM UNLOCKED VEHICLE',
'GRAND THEFT FROM PERSON',
'GRAND THEFT FROM UNLOCKED AUTO',
'GRAND THEFT PICKPOCKET',
'GRAND THEFT PURSESNATCH',
'GRAND THEFT SHOPLIFTING',
'PETTY THEFT FROM UNLOCKED AUTO',
'PETTY THEFT SHOPLIFTING',
'STOLEN BUS',
'THEFT FROM MERCHANT OR LIBRARY',
'THEFT OF CHECKS OR CREDIT CARDS',
'THEFT OF COMPUTERS OR CELL PHONES'
    ] else 'Long' for event in Theft.loc[:,'Descript']]
```

```python
# Conver the 'Hour' to datetime
Theft.loc[:,'Hour'] = pd.to_datetime(Theft.Time).dt.hour

# Count the Theft based on the hour
Theft2 = Theft['Hour'].groupby(Theft['Hour']).agg('count')

# Count the percent of the resolution: 'Unsolved' or 'Solved'
Resolution =  pd.DataFrame(Theft.groupby(['Hour','Resolution2']).agg('count')/Theft.groupby(['Hour']).agg('count'))
Resolution['Solved_Pct'] = Resolution['Address']
Resolution_Pct = list(Resolution[ 'Solved_Pct'])
#print Resolution_Pct

# Split Resolution Percent into Solved_Pct and Unsolved_Pct
Solved_Pct = list()
Unsolved_Pct = list()

for i in xrange(len(Resolution_Pct)):
    if i%2 == 0:
        Solved_Pct.append(Resolution_Pct[i])
    else:
        Unsolved_Pct.append(Resolution_Pct[i])

#print Solved_Pct
#print Unsolved_Pct

# Create the Resolution DataFrame
Resolution_df = pd.DataFrame() 
Resolution_df['Solved_Pct'] = Solved_Pct
Resolution_df['Unsolved_Pct'] = Unsolved_Pct
#print Resolution_df
```

```python
fig = plt.figure() # Create matplotlib figure

ax1 = fig.add_subplot(111) # Create matplotlib axes
ax2 = ax1.twinx() # Create another axes that shares the same x-axis as ax.

# Solved_Pct bar
Resolution_df.Solved_Pct.plot(kind='bar', color='green', alpha = 0.8, ax=ax1, width = 0.4, position=1, label = 'Solved')
# Unsolved_Pct bar
Resolution_df.Unsolved_Pct.plot(kind='bar', color='orange',alpha = 0.6, ax=ax2, width = 0.4, position=0, label = 'Unsolved')

# x axis
ax1.set_xlabel('Time', fontsize = 13)
## Create the % label for x axis
xlabels = ['%i' % i for i in range(24)]
ax1.set_xticklabels(xlabels, rotation=0, fontsize=11)
## Set x axis range
ax1.set_xlim(-1,24)

# Primary y axis
ax1.set_ylabel('Solved %', fontsize= 13)
ax1.tick_params(axis='y', colors='green', labelsize=11)
## Create the % label for Primary y axis
ylabels = ["%i %%" % i for i in range(0, 16, 2)]
ax1.set_yticklabels(ylabels, rotation=0)
## Set primary y axis range
ax1.set_ylim(0,0.14)
## set primary y axis legend location & size
ax1.legend(loc=2, fontsize = 13)

# Secondary y axis
ax2.set_ylabel('Unsolved %', labelpad=5, fontsize= 13)
ax2.tick_params(axis='y', colors='orange', labelsize=11)
## Create the % label for secondary y axis
ylabels2 = ["%i %%" % i for i in range(75, 105, 5)]
ax2.set_yticklabels(ylabels2, rotation=0)
## Set secondary y axis range
ax2.set_ylim(.75,1.00)
## set secondary y axis legend location & size
ax2.legend(loc=1, fontsize=13)
```

```python
# Group the Theft events based on Time_Period
Theft_group_hr = Theft.groupby(['Hour'])
print Theft_group_hr['Hour'].agg('count') / float(len(Theft))
```

```python
# Hourly bar Pct
hour_bar_pct = Theft_group_hr['Hour'].agg('count') / float(len(Theft))       

# Hourly bar hour
hour_bar_hr = range(24)    
   
# Create the bars
barlist = plt.bar(hour_bar_hr, hour_bar_pct, align='center', alpha=0.5)
# Change the bar color
for i in range(len(hour_bar_pct)):    
    barlist[i].set_color((1, 1-hour_bar_pct[i]/max(hour_bar_pct), 0))    

# Find all the peak pts    
peak_pts = []
for i in (range(1,len(hour_bar_pct)-1)):
    if hour_bar_pct[i] > hour_bar_pct[i-1] \
    and hour_bar_pct[i] > hour_bar_pct[i+1]:
        peak_pts.append(i)

# Plot the peak pts        
for i in range(len(peak_pts)):
    plt.scatter(hour_bar_hr[peak_pts[i]], hour_bar_pct[peak_pts[i]], c='magenta', s =500, marker = "*")
        
# Plot the trendline
plt.plot(hour_bar_pct, c='sandybrown', linewidth=1)

# Plot the arrows
plt.arrow(0,0.045,4.5,-0.035, length_includes_head=True,head_width=0.006, head_length=0.6, color = 'orange')
plt.arrow(5,0.015,7.5,0.05, length_includes_head=True,head_width=0.006, head_length=0.6, color = 'red')
plt.arrow(13,0.05,5.5,0.05, length_includes_head=True,head_width=0.006, head_length=0.6, color = 'red')
plt.arrow(19,0.085,4.5,-0.025, length_includes_head=True,head_width=0.006, head_length=0.6, color = 'orange')

# Customize the x ticks
plt.xticks(hour_bar_hr, hour_bar_hr, fontsize=13)

# Customize x label
plt.xlabel('Time', fontsize=15)

# Customize y label
plt.ylabel('Percent', fontsize=15)

# Create the title
plt.title('Hourly Percent of Theft Reports', fontsize=15)

# Grow the y axis
plt.ylim(0,0.105)
# Expand the x axis
plt.xlim(-1, 24)    
    
```

```python
# Hourly bar Pct
hour_bar_pct_pre = list(Theft_group_hr['Hour'].agg('count') / float(len(Theft))) 
## Shift the 0-4 to the end
hour_bar_pct = hour_bar_pct_pre[5:]+hour_bar_pct_pre[:5]

# Hourly bar hours
hour_bar_hr = range(24)

# X axis label
xlabels = range(5, 24) + range(0, 5)

# Create the bars
barlist = plt.bar(hour_bar_hr, hour_bar_pct, align='center', alpha=0.5)
# Change the bar color
for i in range(len(hour_bar_pct)):    
    barlist[i].set_color((1, 1-hour_bar_pct[i]/max(hour_bar_pct), 0))

# Find all the peak pts
peak_pts = list()
for i in (range(1,len(hour_bar_pct)-1)): 
    if hour_bar_pct[i] > hour_bar_pct[i-1] \
    and hour_bar_pct[i] > hour_bar_pct[i+1]:      
        peak_pts.append(i)
        
# Find the max peak pt    
peak_pt_max = list()
for pt in peak_pts:
    if hour_bar_pct[pt] == max([hour_bar_pct[pt] for pt in peak_pts]):
        peak_pt_max.append(pt)       

# Plot the max point
plt.scatter(peak_pt_max[0], hour_bar_pct[peak_pt_max[0]], c='magenta', s =500, marker = "*")

# Plot the trendline
plt.plot(hour_bar_pct, c='sandybrown', linewidth=1)

# Plot the arrows
plt.arrow(0,0.015,13,0.085, length_includes_head=True,head_width=0.006, head_length=0.6, color = 'red')
plt.arrow(13.5,0.1,10,-0.085, length_includes_head=True,head_width=0.006, head_length=0.6, color = 'orange')

# Customize the x ticks values and label
plt.xticks(hour_bar_hr, xlabels, fontsize=14)
plt.xlabel('Time', fontsize=15)

# Customize the y ticks values and label
plt.yticks(fontsize=14)
plt.ylabel('Percent', fontsize=15)

# Create title
plt.title('Hourly Percent of Theft Reports', fontsize=15)

# Grow the y axis 
plt.ylim(0,0.105)
# Expand the x axis 
plt.xlim(-1, 24) 
```
